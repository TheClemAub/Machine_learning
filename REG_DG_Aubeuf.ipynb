### Importation de bibliothèques & Initialisation de nos valeurs

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_samples=50, n_features=1, noise=5) # on génère des données aléatoires grâce à make_regression()
plt.scatter(x, y) # et on affiche les données
```

### Création du modèle

```python
model = linear_model.SGDRegressor(loss='squared_loss', alpha=0.0001, max_iter = 200, learning_rate = 'invscaling') 
# On choisit le type de modèle : régression linéaire via le gradient descendant
# loss='squared loss' : la fonction de régression MSE
# alpha : pas de l'opération
# max_iter : nbr d'itérations du modèle
# learning_rate='invscaling' : schéma d'apprentissage de type invscaling

model.fit(x, y) # On construit notre modèle
prediction = model.predict(x) # On applique notre modèle aux données d'entrée

plt.scatter(x, y)
plt.plot(x, prediction, c='r')
plt.grid()

print('Précision du modèle : ', model.score(x,y)*100, '%') 
# model.score prédit une valeur y' grâce au modèle construit, puis le compare avec y
```

```python
# On récupère notre a et notre b
a = model.coef_
b = model.intercept_

# Puis on arrondit les résultats obtenus
a = round(a[0],2) 
b = round(b[0],2)

# Et on obtient notre fonction calculée
print ('La fonction obtenue est : F = C x', a, '+', b)
```
