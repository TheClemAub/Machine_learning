### Importation de bibliothèques & Initialisation de nos valeurs

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

Pour rappel, la formule de conversion Celsius / Fahrenheit est la suivante : 
$ f = c \times 1.8 + 32 $

```python
celsius = np.array([-20, -10, 0, 10, 20, 25, 30, 38],  dtype=int)
fahrenheit = np.array([-4, 14, 32, 50, 68, 77, 86, 100],  dtype=int)

#celsius, fahrenheit = make_regression(n_samples=8, n_features=1, noise=0) 

# On modifie les dimensions du tableau celsius
celsius = celsius.reshape(-1, 1)

# print(celsius)

plt.xlabel('Degré C°')
plt.ylabel('Degré F')

# On affiche les données
plt.scatter(celsius, fahrenheit)
```

### Création du modèle avec la descente de gradient

```python
model = linear_model.SGDRegressor(loss='squared_loss', alpha=0.0001, max_iter = 600) 
# On choisit le type de modèle : régression linéaire via le gradient descendant
# loss='squared loss' : la fonction de régression MSE
# alpha : pas de l'opération
# max_iter : nbr d'itérations du modèle

model.fit(celsius, fahrenheit) # On construit notre modèle
prediction = model.predict(celsius) # On applique notre modèle aux données d'entrée

plt.scatter(celsius, fahrenheit)
plt.plot(celsius, prediction, c='r')

print('Précision du modèle : ', model.score(celsius,fahrenheit)*100, '%') 
# model.score prédit une valeur y' grâce au modèle construit, puis le compare avec y
```

```python
# On récupère notre a et notre b
a = model.coef_
b = model.intercept_

# Puis on arrondit les résultats obtenus
a = round(a[0],2) 
b = round(b[0],2)

# Et on obtient notre fonction calculée
print ('La fonction obtenue est : F = C x', a, '+', b)
```
